```python
a = gpd.sjoin(reamining_fields,farms, predicate='intersects')
a = a.reset_index(drop=True)
a
```

```python
a.columns
```

```python

geo_idx = 0
for idx, col_n in enumerate(a.columns):
    if col_n=='geometry':
        geo_idx = idx
# Apply this to the geometries that might be invalid
a['geometry'] = a['geometry'].apply(lambda geom: make_valid(geom) if not geom.is_valid else geom)
farms['geometry'] = farms['geometry'].apply(lambda geom: make_valid(geom) if not geom.is_valid else geom)

a['V_O_area_prc'] = a.apply(lambda row: (row[geo_idx].intersection(farms.iloc[row[geo_idx+1]].geometry).area / row[geo_idx].area) * 100, axis=1)
a = a.reset_index(drop=True)
```

```python
idx = a.groupby('fid')['V_O_area_prc'].idxmax()
final_data = a.loc[idx]
final_data = final_data.reset_index(drop=True)
final_data
```
